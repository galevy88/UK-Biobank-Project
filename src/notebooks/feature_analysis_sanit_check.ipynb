```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style for better visualizations
sns.set_style("whitegrid")
%matplotlib inline
```

```python
# Define the file path
file_path = '../data/sanity_check/output/participant_data_by_disease.csv'

# Read the CSV file
df = pd.read_csv(file_path)

# Remove records with at least one null value, including the new column
df = df.dropna()

# Display the first few rows
df.head()
```

```python
# Display basic information about the dataset
print("Dataset Info:")
print(df.info())
print("\nMissing Values:")
print(df.isnull().sum())
print("\nNumber of rows and columns:", df.shape)
```

```python
# Summary statistics for numerical columns
print("Summary Statistics:")
df.describe()
```

```python
# Count the number of samples in each group
print("Group Distribution:")
print(df['label'].value_counts())

# Visualize the group distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='label', data=df)
plt.title('Distribution of Healthy vs Heart Disease')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Select numerical columns (excluding eid, Sex, Smoking status, and label)
numerical_columns = df.select_dtypes(include=[np.number]).columns.drop('eid')

# Set up the figure with subplots
n_cols = 2  # Number of columns for subplots
n_rows = (len(numerical_columns) + n_cols - 1) // n_cols  # Calculate rows needed
plt.figure(figsize=(15, 5 * n_rows))

# Iterate through numerical columns and create histograms
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(n_rows, n_cols, i)
    sns.histplot(data=df, x=column, hue='label', element='step', stat='density', common_norm=False)
    plt.title(f'Distribution of {column} by Group')
    plt.xlabel(column)
    plt.ylabel('Density')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Select numerical columns (excluding eid)
numerical_columns = df.select_dtypes(include=[np.number]).columns.drop('eid')

# Compute Pearson correlation matrix
corr_matrix = df[numerical_columns].corr(method='pearson')

# Create a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0, fmt='.2f')
plt.title('Pearson Correlation Heatmap of Numerical Features')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define dictionary of feature pairs (x_axis: y_axis)
feature_pairs = {
    'Cholesterol | Instance 0': 'LDL direct | Instance 0',
    'Body mass index (BMI) | Instance 0': 'Triglycerides | Instance 0',
    'HDL cholesterol | Instance 0': 'Apolipoprotein A1 | Instance 0'
}

# Set up the figure with subplots
n_pairs = len(feature_pairs)
plt.figure(figsize=(10, 5 * n_pairs))

# Iterate through feature pairs and create scatter plots with regression lines
for i, (x_feature, y_feature) in enumerate(feature_pairs.items(), 1):
    plt.subplot(n_pairs, 1, i)
    sns.regplot(data=df, x=x_feature, y=y_feature, scatter_kws={'alpha':0.5}, hue='label')
    plt.title(f'{x_feature} vs {y_feature} with Regression Line')
    plt.xlabel(x_feature)
    plt.ylabel(y_feature)

plt.tight_layout()
plt.show()
```
