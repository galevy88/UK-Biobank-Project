```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Select numerical columns (excluding eid)
numerical_columns = df.select_dtypes(include=[np.number]).columns.drop('eid')

# Set up the figure with subplots
n_cols = 2  # Number of columns for subplots
n_rows = (len(numerical_columns) + n_cols - 1) // n_cols  # Calculate rows needed
plt.figure(figsize=(15, 5 * n_rows))

# Iterate through numerical columns and create boxplots
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(n_rows, n_cols, i)
    # Plot boxplot
    sns.boxplot(data=df, x='label', y=column, hue='label')

    # Calculate outliers for each group
    outlier_counts = {}
    for label in df['label'].unique():
        group_data = df[df['label'] == label][column]
        Q1 = group_data.quantile(0.25)
        Q3 = group_data.quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = group_data[(group_data < lower_bound) | (group_data > upper_bound)]
        outlier_counts[label] = len(outliers)

    # Update title with outlier counts
    title = f'Boxplot of {column}\nOutliers: Healthy={outlier_counts.get("Healthy", 0)}, Heart Disease={outlier_counts.get("Heart Disease", 0)}'
    plt.title(title)
    plt.xlabel('Group')
    plt.ylabel(column)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from scipy.stats import iqr

# Copy the DataFrame to avoid modifying the original
df_processed = df.copy()

# 1. Transform non-normal features (log transformation for skewed features)
skewed_features = [
    'Lipoprotein A | Instance 0',
    'C-reactive protein | Instance 0',
    'Triglycerides | Instance 0',
    'Glycated haemoglobin (HbA1c) | Instance 0'
]
for feature in skewed_features:
    # Add small constant to avoid log(0) issues, then apply log transformation
    df_processed[feature] = np.log1p(df_processed[feature] + 0.0001)

# 2. Handle outliers (winsorizing: cap at 1.5 * IQR)
numerical_columns = df_processed.select_dtypes(include=[np.number]).columns.drop('eid')
for column in numerical_columns:
    for label in df_processed['label'].unique():
        group_data = df_processed[df_processed['label'] == label][column]
        Q1 = group_data.quantile(0.25)
        Q3 = group_data.quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        # Cap values in the group
        df_processed.loc[(df_processed['label'] == label) & (df_processed[column] < lower_bound), column] = lower_bound
        df_processed.loc[(df_processed['label'] == label) & (df_processed[column] > upper_bound), column] = upper_bound

# 3. Encode categorical features (Sex, Smoking status)
df_processed = pd.get_dummies(df_processed, columns=['Sex', 'Smoking status | Instance 0'], drop_first=True)

# 4. Standardize numerical features
scaler = StandardScaler()
numerical_features = df_processed.select_dtypes(include=[np.number]).columns.drop('eid')
df_processed[numerical_features] = scaler.fit_transform(df_processed[numerical_features])
```

```python
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt

# Prepare features for PCA (exclude eid, label, and Age at recruitment)
X = df_processed.drop(columns=['eid', 'label', 'Age at recruitment'])

# Run PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Create DataFrame for PCA results
pca_df = pd.DataFrame({
    'PC1': X_pca[:, 0],
    'PC2': X_pca[:, 1],
    'label': df_processed['label']
})

# Define custom color palette
palette = {'Healthy': 'blue', 'Heart Disease': 'red'}

# Plot PCA results with custom colors
plt.figure(figsize=(8, 6))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='label', palette=palette, alpha=0.6)
plt.title('PCA: Entire Population (First Two Principal Components)')
plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.2%} variance)')
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.2%} variance)')
plt.show()

# Print explained variance ratio
print("Explained Variance Ratio (Entire Population):", pca.explained_variance_ratio_)
```

```python
# Define subpopulations
obese_smoker_heart = df_processed[
    (df['Body mass index (BMI) | Instance 0'] > 30) &
    (df['Smoking status | Instance 0'] == 'Current') &
    (df['label'] == 'Heart Disease')
]

healthy_non_smoker = df_processed[
    (df['Body mass index (BMI) | Instance 0'].between(18, 24)) &
    (df['Smoking status | Instance 0'] == 'Never') &
    (df['label'] == 'Healthy')
]

# Concatenate subpopulations
subpop_df = pd.concat([obese_smoker_heart, healthy_non_smoker])

# Print number of samples in each group
print(f"Obese Smoker Heart Disease samples: {len(obese_smoker_heart)}")
print(f"Healthy Non-Smoker samples: {len(healthy_non_smoker)}")
```

```python
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt

# Prepare features for PCA (exclude eid, label, and Age at recruitment)
X_subpop = subpop_df.drop(columns=['eid', 'label', 'Age at recruitment'])

# Run PCA
pca_subpop = PCA(n_components=2)
X_subpop_pca = pca_subpop.fit_transform(X_subpop)

# Create DataFrame for PCA results
subpop_pca_df = pd.DataFrame({
    'PC1': X_subpop_pca[:, 0],
    'PC2': X_subpop_pca[:, 1],
    'label': subpop_df['label']
})

palette = {'Healthy': 'blue', 'Heart Disease': 'red'}

# Plot PCA results
plt.figure(figsize=(8, 6))
ax = sns.scatterplot(data=subpop_pca_df, x='PC1', y='PC2', hue='label', palette=palette, alpha=0.6)
plt.axvline(x=2, color='black', linestyle='--', linewidth=1.5, label='PC1 = 2')
plt.axvline(x=-4, color='black', linestyle='--', linewidth=1.5, label='PC1 = -4')
plt.title('PCA: Obese Smoker Heart Disease vs Healthy Non-Smoker')
plt.xlabel(f'PC1 ({pca_subpop.explained_variance_ratio_[0]:.2%} variance)')
plt.ylabel(f'PC2 ({pca_subpop.explained_variance_ratio_[1]:.2%} variance)')
plt.legend()
plt.show()

# Print explained variance ratio
print("Explained Variance Ratio (Subpopulations):", pca_subpop.explained_variance_ratio_)
```

```python
# Create two dataframes based on PC1 thresholds
# First df: samples to the left of PC1 = -4 (PC1 < -4)
df_left_of_minus4 = subpop_pca_df[subpop_pca_df['PC1'] < -4].copy()

# Second df: samples to the right of PC1 = 2 (PC1 > 2)
df_right_of_2 = subpop_pca_df[subpop_pca_df['PC1'] > 2].copy()

# Display information about the filtered dataframes
print("DataFrame 1: Samples to the left of PC1 = -4")
print(f"Number of samples: {len(df_left_of_minus4)}")
print(f"Label distribution:\n{df_left_of_minus4['label'].value_counts()}")
print(f"\nPC1 range: [{df_left_of_minus4['PC1'].min():.2f}, {df_left_of_minus4['PC1'].max():.2f}]")
print(f"PC2 range: [{df_left_of_minus4['PC2'].min():.2f}, {df_left_of_minus4['PC2'].max():.2f}]")

print("\n" + "="*50 + "\n")

print("DataFrame 2: Samples to the right of PC1 = 2")
print(f"Number of samples: {len(df_right_of_2)}")
print(f"Label distribution:\n{df_right_of_2['label'].value_counts()}")
print(f"\nPC1 range: [{df_right_of_2['PC1'].min():.2f}, {df_right_of_2['PC1'].max():.2f}]")
print(f"PC2 range: [{df_right_of_2['PC2'].min():.2f}, {df_right_of_2['PC2'].max():.2f}]")

# Display first few rows of each dataframe
print("\n" + "="*50 + "\n")
print("First few rows of df_left_of_minus4:")
print(df_left_of_minus4.head())

print("\n" + "="*50 + "\n")
print("First few rows of df_right_of_2:")
print(df_right_of_2.head())
```
